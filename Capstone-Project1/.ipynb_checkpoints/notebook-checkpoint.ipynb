### Feature Selection - According to AUC

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.20, random_state=1, stratify=df.obesity_level)

df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)

y_train = df_train.obesity_level.values
y_val = df_val.obesity_level.values

del df_train['obesity_level']
del df_val['obesity_level']
```

```python

len(df_train), len(df_val), len(y_train), len(y_val)
```

```python
columnas = list(df_train.columns)
res_AUC = []
for i in range(len(columnas)):
    model = LogisticRegression(solver='liblinear', random_state=42)
    model.fit(df_train.iloc[:,[i]], y_train)

    y_pred = model.predict_proba(df_val.iloc[:,[i]])

    AUC_metric = roc_auc_score(y_val, y_pred, multi_class='ovr')
    res_AUC.append([columnas[i], AUC_metric])
```

```python
res_AUC = pd.DataFrame(res_AUC, columns=['Feature', 'AUC'])
res_AUC
```

```python
best_AUC = res_AUC.sort_values('AUC', ascending=False)
best_AUC.reset_index(drop=True, inplace=True)
best_AUC
```

```python
features_AUC = [best_AUC.Feature[0]]
limitAUC = best_AUC.AUC[0]
```

```python
for i in range(1, len(best_AUC)):
    features_AUC.append(best_AUC.iloc[i, 0])
    datosZ = df_train.loc[:, features_AUC]
    modeloZ = LogisticRegression(solver='liblinear', random_state=42)
    modeloZ.fit(datosZ, y_train)
    
    prediccionesZ = modeloZ.predict_proba(df_val.loc[:, features_AUC])
    
    AUC_metric = roc_auc_score(y_val, prediccionesZ, multi_class='ovr')

    print(features_AUC, AUC_metric)
    
    if AUC_metric > limitAUC:
        limitAUC = AUC_metric
    else:
        del features_AUC[-1]

print(len(features_AUC))
print(limitAUC)
```

### Model Selection

Logistic Regression was implemented already, now we are going to implement XGBoost.

* All features

```python
train_dicts = df_train.to_dict(orient='records')
```

```python
dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)
```

```python
dv.feature_names_
```

```python
val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
features = dv.feature_names_
```

```python
dtrain = xgb.DMatrix(X_train, label = y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
xgb_params = {
    'eta': 0.3,
    'max_depth': 6,
    'min_child_weight':1,

    'objective': 'multi:softmax',
    'num_class': 7,
    'eval_metric': 'auc',
    'nthread': 8,

    'seed': 42,
    'verbosity': 1
}

model = xgb.train(xgb_params, dtrain, num_boost_round = 200 )
```

```python
y_pred = model.predict(dval)
```

```python
probabilities = model.predict(dval, output_margin=True)
probabilities = np.exp(probabilities) / np.sum(np.exp(probabilities), axis=1, keepdims=True)
```

```python
roc_auc_score(y_val, probabilities, multi_class='ovr')
```

```python
accuracy_score(y_val, y_pred)
```

### Parameter Tuning

```python
train_dicts = df_train.to_dict(orient='records')
```

```python
dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)
```

```python
dv.feature_names_
```

```python
features = dv.feature_names_
```

```python
dtrain = xgb.DMatrix(X_train, label = y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
watchlist = [(dtrain, 'train'), (dval, 'val')]
```

```python
def parse_xgb_output(output):
    results = []
    
    for line in output.stdout.strip().split('\n'):
        it_line, train_line, val_line = line.split('\t')

        it = int(it_line.strip('[]'))
        train = float(train_line.split(':')[1])
        val = float(val_line.split(':')[1])

        results.append((it, train, val))


    columns = ['num_iter','train_auc','val_auc']
    df_results = pd.DataFrame(results, columns=columns)
    return df_results
```
